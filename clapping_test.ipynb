```python
from audio import *
```

```python
fs = 44100
sd.default.samplerate = fs
T = 3
print("Recording...")
clap = sd.rec(T * fs)
sd.wait()
print("Finished recording")
```

```python
sd.play(clap)

time = np.linspace(0,T, T * fs)
y_max_index= np.where(clap == np.amax(clap))[0]

max_x= time[y_max_index]

x= time[::100]
y= clap[::100]

plt.figure(1)
plt.title("Clap Bedroom")
plt.xlabel('Time, s')
plt.plot(x, y)
plt.ylim(-np.amax(y)*1.1, np.amax(y)*1.1)
plt.grid()
plt.savefig('plots/claps/bedroom.jpeg')
plt.show()
```
