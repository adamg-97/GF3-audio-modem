```python
from audio import *
```

```python
fs = 44100
sd.default.samplerate = fs
T = 3
print("Recording...")
clap = sd.rec(T * fs)
sd.wait()
print("Finished recording")
```

```python
sd.play(clap)
time = np.linspace(0,T*fs, T * fs)

x= time[::50]
y= clap[::50]

y_max_index= np.where(clap == np.amax(clap))[0]
print(np.amax(clap))

max_x= time[y_max_index]



plt.figure(1)
plt.title("Clap Bedroom")
plt.xlabel('Time, s')
plt.plot(x, y)
plt.xlim(max_x-1000,max_x+10000)
plt.ylim(-np.amax(y)*1.1, np.amax(y)*1.1)
plt.grid()
plt.savefig('plots/claps/bedroom.jpeg')
plt.show()
```
